## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
line_subtotal = client_df['unit_price'] * client_df['qty']
client_df["line_subtotal"] = line_subtotal

client_df[['unit_price','qty','line_subtotal']].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

# calculate total weight of order.
total_weight = client_df['unit_weight'] * client_df['qty']
client_df['total_weight'] = total_weight
client_df


client_df['shipping_price'] = client_df['total_weight'].apply(lambda x: x * 7 if x > 50 else x * 10)

client_df[['unit_price','unit_weight','qty','total_weight','shipping_price']].head(3)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

client_df['line_price'] = ((client_df['line_subtotal'] + client_df['shipping_price']) * 1.0925).round(2)

client_df[['line_subtotal','shipping_price', 'line_price']].head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

line_cost = (client_df['unit_cost'] * client_df['qty']) + client_df['shipping_price']
client_df['line_cost'] = line_cost
client_df.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price

line_profit = client_df['line_price'] - client_df['line_cost']
client_df['line_profit'] = line_profit.round(2)
client_df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

top5_clients = [33615, 66037, 46820, 38378, 24741]

for client_id in top5_clients:  
    client_id_isolated = client_df[client_df.client_id == client_id]
    top5_total_price = client_id_isolated['line_price'].sum()
    print(f" {client_id} : ${top5_total_price:,.2f}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

top5_df = client_df.loc[client_df['client_id'].isin(top5_clients)]
summary_top5 = top5_df[['client_id', 'qty', 'shipping_price', 'line_price', 'line_cost', 'line_profit']]
summary_top5

Summary = summary_top5.groupby(['client_id']).sum()
Summary.sort_values('line_profit', ascending=False)

```

```python
# Format the data and rename the columns
# to names suitable for presentation.

Summary.rename(columns={'qty':'Units','shipping_price':'Shipping','line_price':'Total Revenue', 'line_cost':'Total Cost','line_profit':'Total Profit'}, inplace=True)
Summary
```

```python
# Currency should be in millions of dollars.

def format_to_millions(series):
    return series.apply(lambda x: "${:.2f}M".format(x / 1000000))


Summary[['Shipping', 'Total Revenue', 'Total Cost', 'Total Profit']] = Summary[['Shipping', 'Total Revenue', 'Total Cost', 'Total Profit']].apply(format_to_millions)

Summary.head()
```
